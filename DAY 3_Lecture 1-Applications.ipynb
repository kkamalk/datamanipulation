```python
# Group the data by position
position_groups = nba2.groupby("Position")

# Iterate through each group and calculate the median salary since it aint affected by outliers
position_medians = {position: group["Salary"].median() for position, group in position_groups}

# Iterate through each row in the dataset and fill in missing salary values with the median for that position
for index, row in nba2.iterrows():
    if pd.isna(row["Salary"]):
        nba2.at[index, "Salary"] = position_medians[row["Position"]]
```

```python
k = []
value_median = []
value_mean = []
for val in nba3[nba3['Salary'].isna()]['Team'].unique():
    k.append(val) 
    value_median.append(nba3[nba3['Team']==val]['Salary'].median())
    value_mean.append(nba3[nba3['Team']==val]['Salary'].mean())
d = dict(zip(k,value_median))
e = dict(zip(k,value_mean))
```

##  <font color='Dark Orange'>Learn More for meta-data</font>


```python
s = [["meta", "meta", "data", "data", "time", "time", "joke", "joke"],
    ["one", "two", "one", "two", "one", "two", "one", "two"]]
```

```python
# MultiIndex.from_arrays
pd.MultiIndex.from_arrays(s,names=["1st", "2nd"])
```

```python
# MultiIndex.from_tuples
s_tuple = list(zip(*s))
s_tuple
index_1= pd.MultiIndex.from_tuples(s_tuple, names=["1st", "2nd"])
```

```python
index_1
```

```python
# MultiIndex.from_product
pro = [["meta", "data", "time", "joke"], ["one", "two"]]
pd.MultiIndex.from_product(pro, names=["1st", "2nd"])
```

```python
# MultiIndex.from_frame
df = pd.DataFrame([["meta", "one"], ["meta", "two"], ["data", "one"], ["data", "two"],['time','one'],['time','two'],
                   ['joke','one'],['joke','two']], columns=["1st", "2nd"])
pd.MultiIndex.from_frame(df)
```
