### <font color='Dark orange'> 1) Aggregation</font> 

```python
import numpy as np
```

```python
grp1 = car.groupby('Car_Name')
grp1.aggregate(np.sum)
```

```python
grp2 = car.groupby(['Car_Name','Fuel_Type'])
grp2.aggregate(np.sum)
```

```python
## applying a list of different functions to one column
grp3 = car.groupby(['Car_Name','Fuel_Type'])

grp3['Kms_Driven'].agg([np.sum, np.min, np.max, np.mean])
```

```python
# using different function to different columns
grp4 = car.groupby(['Car_Name','Fuel_Type'])

grp4.agg({'Year':'min','Selling_Price':'mean','Kms_Driven': 'max'})
```

**Transformation:** </br> 
Transformation is a process in which we perform some group-specific computations and return a like-indexed. Transform method returns an object that is indexed the same (same size) as the one being grouped. The transform function must: 
 

* Return a result that is either the same size as the group chunk
* Operate column-by-column on the group chunk
* Not perform in-place operations on the group chunk.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grp5 = car.groupby(['Car_Name','Fuel_Type'])

grp5.transform(lambda x:(x-x.min())/x.max()) # min-max scaling(a kind of standardization)
```

## Difference between Apply and Transform strategies!

```python
car_sample = car.drop(['Owner'],axis=1)
car_sample1 = car_sample[car_sample['Car_Name'].str.lower().str.startswith('s')]
```

```python
car_sample1
```

```python
# we sold the car and the price at present is described by Present_Price
car_sample1['New'] = car_sample1.apply(lambda row: row.Selling_Price-row.Present_Price,axis=1)
```

```python
car_sample1
```

```python
car_sample1['New_trans'] = car_sample1.transform(lambda row: row.Selling_Price-row.Present_Price)
```

```python
car_sample1['New_trans'] = car_sample1.transform(lambda row: row.Selling_Price-row.Present_Price,axis=1)
```

The apply function sends a whole copy of the dataframe to work upon so we can manipulate all the rows or columns simultaneously.

#### What do you think which car(s) is/are the most profitable?

```python
car_name_profit = car_sample.copy()
```

```python
# we sold the car and the price at present is described by Present_Price
# relative error = (estimate-actual)/actual
car_name_profit['Loss%']=car_name_profit.apply(lambda x: abs((x['Selling_Price']-x['Present_Price'])/x['Present_Price'])*100,axis=1)
```

### Another example is on netflix:

```python
netflix = pd.read_excel('netflix.xlsx')
```

```python
netflix.head()
```

```python
netflix.info()
```

```python
netflix.dropna(inplace=True) # drop null values
netflix.drop_duplicates(inplace=True)
```

```python
grp7=netflix.groupby('release year').median()
```

```python
import matplotlib.pyplot as plt
plt.plot(grp7.index,netflix.groupby('release year')['user rating score'].agg(np.mean))
plt.xlabel('year of release')
plt.ylabel('mean of rating')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# To visualize the seanality & trendy & noisy kindness we utilize the seasonal_decompose()

plt.figure(num=None, figsize=(50, 20), dpi=80, facecolor='w', edgecolor='k')
result=seasonal_decompose(netflix.groupby('release year')['user rating score'].agg(np.mean), model='multiplicative', period=1)
result_plot=result.plot()
```
