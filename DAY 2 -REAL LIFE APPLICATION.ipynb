```python
import matplotlib.pyplot as plt #plot package
import pandasql as ps #sql package
import wordcloud #will use for the word cloud plot
from wordcloud import WordCloud, STOPWORDS # optional to filter out the stopwords
#To build a wordcloud, we have to remove NULL values first:
aff_withoutna= (scopus['Author Keywords']+scopus['Index Keywords']).fillna(value="")
#aff_withoutna= scopus['Source title'].fillna(value="")
#Now let's add a string value instead to make our Series clean:
word_string=" ".join(aff_withoutna.str.lower())
#Plot:
plt.figure(figsize=(15,15))
wc = WordCloud(background_color="pink", stopwords = STOPWORDS, max_words=2000, max_font_size= 300,  width=1600, height=800)
wc.generate(word_string)
plt.imshow(wc.recolor( colormap= 'plasma' , random_state=17), interpolation="bilinear")
plt.axis('off')
```

```python
# Because of our scenario! Just need to see the readable article name, year and reachable links
uni_pubmed=Pubmed_drop[['Title', 'Publication Year', 'DOI']]
uni_scopus=Scopus_drop[['Title', 'Year','DOI']]
```

```python
# Create the union of these datasets
union=uni_scopus.append(uni_pubmed)
```

```python
union
```

```python
union_data=union.drop_duplicates(subset='DOI', keep='first',inplace=False,ignore_index=True)
```

```python
union_data
```

```python
merged_PPI1=union_data.loc[(union_data['Title'].str.contains("Proton Pump")) |(union_data['Title'].str.contains("proton pump"))|
                           (union_data['Title'].str.contains("Proton pump"))]
merged_PPI2=union_data.loc[(union_data['Title'].str.contains("Acidity")) |(union_data['Title'].str.contains("acidity"))]
merged_PPI3=union_data.loc[(union_data['Title'].str.contains("vacuolar ATPase")) |(union_data['Title'].str.contains("Vacuolar ATPase"))]
merged_PPI4=union_data.loc[(union_data['Title'].str.contains("vacuolar H+ ATPase")) |(union_data['Title'].str.contains("Vacuolar H+ ATPase"))]
merged_PPI5=union_data.loc[(union_data['Title'].str.contains("V-ATPase")) |(union_data['Title'].str.contains("V-ATPase"))]
merged_PPI6=union_data.loc[(union_data['Title'].str.contains("proton pump")) & (union_data['Title'].str.contains("therapeutic"))]
merged_PPI7=union_data.loc[(union_data['Title'].str.contains("proton pump inhibitor")) |(union_data['Title'].str.contains("Proton Pump Inhibitor"))|
                           (union_data['Title'].str.contains("Proton pump inhibitor"))]
```

```python
merged_data=merged_PPI1.append([merged_PPI2,merged_PPI4,merged_PPI5,merged_PPI6,merged_PPI7])
```
