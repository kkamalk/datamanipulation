```python
# Let's say you want to find the mean and median of the "Page count" column
mean_value = df2["Page count"].mean()
median_value = df2["Page count"].median()

print("Mean value of Page count:", mean_value)
print("Median value of Page count:", median_value)
```

```python
# Group the data by position
position_groups = df2.groupby("DOI")

# Iterate through each group and calculate the median  since it aint affected by outliers
position_medians = {position: group["Page count"].median() for position, group in position_groups}

# Iterate through each row in the dataset and fill in missing salary values with the median for that position
for index, row in df2.iterrows():
    if pd.isna(row["Page count"]):
        df2.at[index, "Page count"] = position_medians[row["DOI"]]
```

```python
data=pd.read_csv('/content/csv-Machinelea-set.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
publications_per_year = data["Publication Year"].value_counts().sort_index()

# plot the bar chart
plt.bar(publications_per_year.index, publications_per_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Publications")
plt.title("Number of Publications per Year")
plt.show()
```

```python
authors_per_publication = data.groupby("Publication Year")["Authors"].apply(lambda x: x.str.count(";") + 1)

# plot the scatter plot
plt.scatter(authors_per_publication.index, authors_per_publication.values)
plt.xlabel("Year")
plt.ylabel("Number of Authors")
plt.title("Number of Authors per Publication")
plt.show()
```

```python
from wordcloud import WordCloud
titles = " ".join(data["Title"].values)

wordcloud = WordCloud().generate(titles)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Titles")
plt.show()
```
