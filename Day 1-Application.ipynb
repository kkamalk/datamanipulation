```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import io
```

```python
# data for gene expression:colorectal cancer patients data
url2="https://raw.githubusercontent.com/cs109/2014_data/master/exprs_GSE5859.csv"
s2=requests.get(url2).content
c2=pd.read_csv(io.StringIO(s2.decode('utf-8')))
```

```python
c2.head(5)
```

```python
c2.sample(5)
```

```python
c2.columns
```

```python
c2.rename(columns={'Unnamed: 0':'ID_ref'},inplace=True)
```

```python
c2.head(3)
```

```python
# getting the transpose of the data to align with patients' data
df_gene = c2.transpose()
# Let's check the first 10 data
df_gene.head(10)
```

```python
# to use the CPU time effectively, let's shrink the data
df_gene_sample = df_gene.iloc[:,0:70]
```

```python
df_gene_sample.head(5)
```

```python
# use first content row as a column name:
col_names = df_gene_sample.iloc[0].tolist()
#col_names2 = c2['ID_ref'][0:70]
df_gene_sample.columns = col_names
```

```python
df_gene_sample.head(1)
```

```python
# remove the redundant fisrt row:
df_gene_sample = df_gene_sample.drop(axis=0,index='ID_ref')
```

```python
df_gene_sample.head(2)
```

```python
# add ID_ref (as a first colum) by copyin index
df_gene_sample.insert(loc=0,column='ID_ref',value =df_gene_sample.index)
```

```python
df_gene_sample.head(5)
```

```python
## UTILIZE df.info  check the dtyes of values!!!!
```

```python
# reset index
df_gene_sample = df_gene_sample.reset_index(drop=True)

# convert to numerical
features_gene_number = df_gene_sample.columns.tolist()[1:]
```

```python
df_gene_sample[features_gene_number] = df_gene_sample[features_gene_number].astype(float)
```

```python
df_gene_sample
```

```python
# to see main statistics of the data
df_gene_sample.describe()
```

```python
corr_pearson = df_gene_sample[features_gene_number].corr(method = 'pearson')

import seaborn as sns

# getting the Upper triagular corr-matrix
trimask = np.triu(np.ones_like(corr_pearson,dtype =bool))
fig = plt.figure(figsize=(12,10))
# using the upper triangular matrix as mask
sns.heatmap(corr_pearson, cmap='RdYlGn_r', mask=trimask|(np.abs(corr_pearson)>=0.5),annot=True)
```

```python
## It is a good example but we do not know everything yet. we will see all the thing during our lectures:
corr_pearson1= corr_pearson.mask(trimask)

# convert matrix in data frame
df_corr = pd.DataFrame(corr_pearson1).melt(var_name='x2', 
                                          value_name='corr_p',
                                          ignore_index=False)
df_corr.insert(loc=0, column='x1', value=df_corr.index)

# remove redundancies
df_corr = df_corr[~df_corr.corr_p.isna()]

# reset index
df_corr = df_corr.reset_index(drop=True)
```

```python
df_corr
```

```python
# top 10 positive correlations
df_corr.nlargest(10, columns='corr_p')
```

```python
from scipy import stats
# annova test statistics: testing null hypothesis that is the selected tickets 
#have the equal importance
F, p = stats.f_oneway(c3[c3['Ticket'].str.contains('A/5')].Survived,
                      c3[c3['Ticket'].str.contains('CA ')].Survived,
                      c3[c3['Ticket'].str.contains('PC ')].Survived)
F, p
```

```python
## p<= 0.05 we reject the null hypo== at least one of the 
# selected tickets is haveing an impotance of the survived
```
